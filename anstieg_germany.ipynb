Eine der Hauptursachen für den Anstieg der Solarstromerzeugung in Deutschland ab 2009 war die Einführung des Erneuerbare-Energien-Gesetzes (EEG) im Jahr 2000 und seine Novellierung im Jahr 2004. Das EEG war ein wegweisendes Gesetz, das Investitionen in erneuerbare Energien wie Solarenergie, Windenergie und Biomasse förderte, indem es eine Einspeisevergütung für Strom aus erneuerbaren Quellen garantierte. Diese Einspeisevergütung war höher als der Marktpreis für Strom und machte den Betrieb von Solaranlagen wirtschaftlich rentabel.  
Darüber hinaus führte die Entwicklung und Produktion von Solartechnologie zu sinkenden Preisen und machte die Installation von Solaranlagen auch für Privatpersonen und Unternehmen erschwinglicher.

```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Read the data
hydropower_consumption_df = pd.read_csv('Data/All/Hydropower/hydropower-consumption.csv')
```

```python
# Print the first 5 rows of the dataframe
hydropower_consumption_df.head()
```

```python
# filter rows with entity as 'Switzerland', 'Germany', 'United Kingdom', 'Spain'
hydropower_consumption_countries_df = hydropower_consumption_df[(hydropower_consumption_df['Entity'] == 'Switzerland') | (hydropower_consumption_df['Entity'] == 'Germany') | (hydropower_consumption_df['Entity'] == 'United Kingdom') | (hydropower_consumption_df['Entity'] == 'Spain')]
hydropower_consumption_countries_df
```

```python
# plot the data
plt.figure(figsize=(15, 10))
sns.lineplot(x='Year', y='Electricity from hydro (TWh)', hue='Entity', data=hydropower_consumption_countries_df)
plt.title('Hydropower Consumption in Switzerland, Germany, United Kingdom and Spain')
plt.xlabel('Year')
plt.ylabel('Hydropower Consumption (TWh)')
plt.show()
```
